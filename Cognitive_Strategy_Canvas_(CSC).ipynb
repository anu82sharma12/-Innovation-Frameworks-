```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt

departments = ["Finance","HR","Ops","Sales","R&D"]
ai_maturity = np.random.randint(40,90,len(departments))
roi_potential = np.random.randint(10,100,len(departments))
plt.scatter(ai_maturity, roi_potential, color='teal')
for i, dept in enumerate(departments):
    plt.text(ai_maturity[i]+1, roi_potential[i]+1, dept)
plt.xlabel("AI Maturity (%)"); plt.ylabel("ROI Potential (%)")
plt.title("Cognitive Strategy Canvas â€“ AI Maturity vs ROI Potential")
plt.grid(True); plt.show()
```
