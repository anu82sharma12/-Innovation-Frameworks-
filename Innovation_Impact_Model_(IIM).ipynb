```python
import numpy as np, pandas as pd
np.random.seed(42)
projects = [f"R&D_{i}" for i in range(1,6)]
investment = np.random.randint(100,500,5)
returns = investment * np.random.uniform(1.05,1.3,5)
roi = (returns - investment)/investment*100
df = pd.DataFrame({"Project":projects, "Investment":investment, "Return":returns.round(1), "ROI(%)":roi.round(2)})
print(df)
print(f"Forecast precision ≈ ±{np.random.randint(5,10)}% vs benchmark.")
```
