```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt

domains = ["Data","Model","Deployment","Compliance"]
baseline = np.random.randint(60,90,4)
post_audit = baseline + np.random.randint(-5,5,4)
improvement = baseline - post_audit
df = pd.DataFrame({"Domain":domains, "Baseline_Risk":baseline, "Post_Audit_Risk":post_audit})
plt.bar(domains, improvement, color="darkorange")
plt.title("Ethical AI Governance Model â€“ Risk Reduction by Domain")
plt.ylabel("Risk Reduction (%)");
plt.show()
```
