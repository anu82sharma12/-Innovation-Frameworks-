```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
stages = ["Assessment","Pilot","Integration","Rollout"]
baseline_time = np.random.randint(15,40,len(stages))
optimized = baseline_time - np.random.randint(5,15,len(stages))
plt.plot(stages, baseline_time, 'r--o', label="Legacy")
plt.plot(stages, optimized, 'g-o', label="Quantum-Ready")
plt.title("Quantum Adoption Pathway  Implementation Time Reduction")
plt.ylabel("Days"); plt.legend(); plt.show()
```
