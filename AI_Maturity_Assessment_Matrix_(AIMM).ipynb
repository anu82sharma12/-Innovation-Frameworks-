```python
import numpy as np, pandas as pd
indicators = [f"Metric_{i}" for i in range(1,21)]
baseline = np.random.randint(50,90,20)
ai_ready = baseline + np.random.randint(-5,10,20)
df = pd.DataFrame({"Indicator":indicators, "Baseline":baseline, "AI_Ready":ai_ready})
df["Improvement_%"] = (df["AI_Ready"]-df["Baseline"]).round(2)
print(df.head())
print(f"Avg transformation clarity gain ≈ {df['Improvement_%'].mean():.1f}%")
```
